```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('datasets/KMeans_Mall_sale.csv')
data.head()
```

```python
input_case1 = data[['Annual Income (k$)', 'Spending Score (1-100)']]
input_case2 = data[['Age', 'Spending Score (1-100)']]
```

<B> Considering columns Annual Income (k$) and Spending Score (1-100)

```python
input_case1 = data[['Annual Income (k$)', 'Spending Score (1-100)']]
from sklearn.cluster import KMeans
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(input_case1)
    sse.append(km.inertia_)
plt.figure(figsize= (7, 6))
plt.xlabel('K-Range')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
km = KMeans(n_clusters=5)
km.fit(input_case1)
input_case1['cluster'] = km.labels_
df1 = input_case1[input_case1.cluster==0]
df2 = input_case1[input_case1.cluster==1]
df3 = input_case1[input_case1.cluster==2]
df4 = input_case1[input_case1.cluster==3]
df5 = input_case1[input_case1.cluster==4]
plt.figure(figsize= (7, 6))
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],s = 100, color='green', label = 'cluster 1')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],s = 100, color='red', label = 'cluster 2')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],s = 100, color='blue', label = 'cluster 3')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'],s = 100, color='purple', label = 'cluster 4')
plt.scatter(df5['Annual Income (k$)'],df5['Spending Score (1-100)'],s = 100, color='yellow', label = 'cluster 5')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s = 300, color='black',marker = '*' ,label='centroid')
plt.legend()
```

```python
from sklearn.cluster import KMeans
sse_new = []
k_rng = range(1,11)
for k in k_rng:
    km_new = KMeans(n_clusters=k)
    km_new.fit(input_case2)
    sse_new.append(km_new.inertia_)
plt.figure(figsize= (7, 6))
plt.xlabel('K-Range')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse_new)
```
